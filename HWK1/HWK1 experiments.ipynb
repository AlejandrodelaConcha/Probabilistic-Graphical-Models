```python
import numpy as np 
from classifiers import *
import matplotlib.pyplot as plt 
from sklearn.metrics import confusion_matrix
```

## Loading Datasets

```python
train_A=np.loadtxt('classification_data_HWK1\classificationA.train')  
test_A=np.loadtxt('classification_data_HWK1\classificationA.test')

x_train_A,y_train_A=train_A[:,:2],train_A[:,2]
x_test_A,y_test_A=test_A[:,:2],test_A[:,2]

train_B=np.loadtxt('classification_data_HWK1\classificationB.train')  
test_B=np.loadtxt('classification_data_HWK1\classificationB.test')

x_train_B,y_train_B=train_B[:,:2],train_B[:,2]
x_test_B,y_test_B=test_B[:,:2],test_B[:,2]

train_C=np.loadtxt('classification_data_HWK1\classificationC.train')  
test_C=np.loadtxt('classification_data_HWK1\classificationC.test')

x_train_C,y_train_C=train_C[:,:2],train_C[:,2]
x_test_C,y_test_C=test_C[:,:2],test_C[:,2]
```

## LDA

```python
LDA_A=LDA()
LDA_A.fit(x_train_A,y_train_A)
y_predict_LDA_A=LDA_A.predict(x_test_A)
print("Confusion_matrix: \n"+str(confusion_matrix(y_predict_LDA_A,y_test_A)/len(y_predict_LDA_A)))
LDA_A.plot(x_test_A,y_test_A)
```

```python
LDA_B=LDA()
LDA_B.fit(x_train_B,y_train_B)
y_predict_LDA_B=LDA_B.predict(x_test_B)
print("Confusion_matrix: \n"+str(confusion_matrix(y_predict_LDA_B,y_test_B)/len(y_predict_LDA_B)))
LDA_B.plot(x_test_B,y_test_B)
```

```python
LDA_C=LDA()
LDA_C.fit(x_train_C,y_train_C)
y_predict_LDA_C=LDA_C.predict(x_test_C)
print("Confusion_matrix: \n"+str(confusion_matrix(y_predict_LDA_C,y_test_C)/len(y_predict_LDA_C)))
LDA_C.plot(x_test_C,y_predict_LDA_C)
```

## QDA

```python
QDA_A=QDA()
QDA_A.fit(x_train_A,y_train_A)
y_predict_QDA_A=QDA_A.predict(x_test_A)
print("Confusion_matrix: \n"+str(confusion_matrix(y_predict_QDA_A,y_test_A)/len(y_predict_QDA_A)))
QDA_A.plot(x_test_A,y_test_A)
```

```python
QDA_B=QDA()
QDA_B.fit(x_train_B,y_train_B)
y_predict_QDA_B=QDA_B.predict(x_test_B)
print("Confusion_matrix: \n"+str(confusion_matrix(y_predict_QDA_B,y_test_B)/len(y_predict_QDA_B)))
QDA_B.plot(x_test_B,y_test_B)
```

```python
QDA_C=QDA()
QDA_C.fit(x_train_C,y_train_C)
y_predict_QDA_C=QDA_C.predict(x_test_C)
print("Confusion_matrix: \n"+str(confusion_matrix(y_predict_QDA_C,y_test_C)/len(y_predict_QDA_C)))
QDA_C.plot(x_test_C,y_test_C)
```
