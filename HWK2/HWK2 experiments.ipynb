```python
import numpy as np 
from unsupervised_classifiers import *
import matplotlib.pyplot as plt 
from sklearn.metrics import confusion_matrix
```

## Loading datasets

```python
x_train=np.loadtxt('classification_data_HWK2\EMGaussian.data')  
x_test=np.loadtxt('classification_data_HWK2\EMGaussian.test')
```

## K-means

```python
k_means_=K_means(K=4)
y_train=k_means_.fit(x_train)
y_test=k_means_.predict(x_test)
k_means_.plot(x_train,y_train)
```

## Isotropic Gaussian Mixture 

```python
gaussian_mixture_isotropic_=Gaussian_mixture(K=4,isotropic=True)
y_train=gaussian_mixture_isotropic_.fit(x_train)
gaussian_mixture_isotropic_.plot(x_train,y_train)
```

## Gaussian Mixture

```python
gaussian_mixture_=Gaussian_mixture(K=4)
y_train=gaussian_mixture_.fit(x_train)
gaussian_mixture_.plot(x_train,y_train)
```
